## 1. Feature Engineering: Extracción de Franjas Horarias

Clasificamos la hora de contacto en Mañana (8am-12pm), Tarde (12pm-5pm) y Noche (5pm-9pm) para definir la variable temporal.

```python
def get_time_slot(hora_str):
    if pd.isnull(hora_str):
        return 'Mañana'
    # Extraer la hora y convertir a entero
    hora = int(hora_str.split(':')[0])
    if 8 <= hora < 12:
        return 'Mañana'
    elif 12 <= hora < 17:
        return 'Tarde'
    else:
        return 'Noche'

# Aplicar transformación
contactos['momento_contacto'] = contactos['hora_contacto'].apply(get_time_slot)
print(contactos['momento_contacto'].value_counts())
```
