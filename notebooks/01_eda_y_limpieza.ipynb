# 01. Análisis Exploratorio de Datos (EDA) y Limpieza - Mibanco NBA

Este notebook sirve para la experimentación sucia del equipo sobre los tres datasets entregados por Mibanco:
- Clientes (`01_Tabla_de_Clientes.csv`)
- Créditos (`02_Tabla_de_Crditos.csv`)
- Historial de contactos (`03_Tabla_contactos.csv`)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Rutas relativas a data/01_raw
RAW_DIR = "../data/01_raw"

print("Entorno inicializado. Cargando datasets de prueba...")
```

## 1. Carga Preliminar de Datos

```python
clientes = pd.read_csv(os.path.join(RAW_DIR, "01_Tabla_de_Clientes.csv"))
creditos = pd.read_csv(os.path.join(RAW_DIR, "02_Tabla_de_Crditos.csv"))
contactos = pd.read_csv(os.path.join(RAW_DIR, "03_Tabla_contactos.csv"))
```

## 2. Inspección General e Imputación de Nulos

```python
print("Nulos en Clientes:")
print(clientes.isnull().sum())

print("\nNulos en Créditos:")
print(creditos.isnull().sum())

print("\nNulos en Contactos:")
print(contactos.isnull().sum())
```

## 3. Distribución de Variables Clave (Frecuencia de Contactos, Canales y Horarios)

```python
# Analizar la distribución del canal de contacto histórico
plt.figure(figsize=(8, 5))
sns.countplot(data=contactos, x='canal_contacto', order=contactos['canal_contacto'].value_counts().index)
plt.title('Distribución de Contactos por Canal Histórico')
plt.show()
```
