```python
import os
from pathlib import Path
import pandas as pd
from concurrent.futures import ProcessPoolExecutor
from concurrent.futures import ThreadPoolExecutor

# Configuración de rutas
INPUT_DIR = "./inputs"
OUTPUT_DIR = "./inputs_transformed"

def convert_excel_to_csv(excel_path):
    try:
        # Reemplazar extensión por .csv
        csv_name = Path(excel_path).stem + ".csv"
        csv_path = os.path.join(OUTPUT_DIR, csv_name)
        
        # Leer todas las pestañas si tiene varias, o solo la primera (ajustar según necesidad)
        # Nota: 'openpyxl' es el motor estándar y eficiente para .xlsx
        df = pd.read_excel(excel_path, engine='openpyxl')
        
        # Guardar a CSV de manera rápida
        df.to_csv(csv_path, index=False, encoding='utf-8')
        return f"✅ Convertido: {os.path.basename(excel_path)}"
    except Exception as e:
        return f"❌ Error en {os.path.basename(excel_path)}: {str(e)}"

def batch_conversion():
    # Asegurar que la carpeta de salida exista
    os.makedirs(OUTPUT_DIR, exist_ok=True)
    
    # Listar todos los archivos Excel (.xlsx o .xls)
    excel_files = [
        os.path.join(INPUT_DIR, f) for f in os.listdir(INPUT_DIR) 
        if f.endswith(('.xlsx', '.xls')) and not f.startswith('~$')
    ]
    
    if not excel_files:
        print("No se encontraron archivos Excel.")
        return

    print(head_msg := f"Iniciando conversión en paralelo de {len(excel_files)} archivos...")
    
    # Ejecución en paralelo usando los cores disponibles
    with ProcessPoolExecutor() as executor:
        results = executor.map(convert_excel_to_csv, excel_files)
        
        for result in results:
            print(result)

if __name__ == "__main__":
    # En entornos interactivos (Jupyter) ProcessPoolExecutor puede fallar;
    # usar ThreadPoolExecutor en su lugar para evitar BrokenProcessPool.
    globals()['ProcessPoolExecutor'] = ThreadPoolExecutor

    batch_conversion()
```
